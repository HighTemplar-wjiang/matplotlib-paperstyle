```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Set rcParams
mpl.style.use("paper")
```

```python
# Example
import numpy as np
x = np.arange(0, 4*np.pi, 0.01)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y, label="Sin")
ax.set_title("Sin wave")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.legend()

fig.savefig("./test.pdf")
```
